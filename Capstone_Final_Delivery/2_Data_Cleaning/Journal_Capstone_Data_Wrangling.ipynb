```python
import pandas as pd
```

Data acquried from web scraping the journal information from https://exaly.com/

```python
# Read csv into a pandas dataset
raw_df = pd.read_csv('Master_literature_impact_factor_file_11_15.csv')
raw_df.head()
```

```python
# Rename columns to facilitate understanding
raw_df_rn = raw_df.rename(index=str, columns={"citation_number":"Total_number_of_citation", "cited_by":"Citation_most_cited_article", "year":"Year_of_most_cited_article"})
raw_df_rn.head()
```

```python
# Drop columns that are not interested in 
df_target = raw_df_rn.drop(['jornal_url', 'most_cited'], axis=1)
df_target.head()
```

```python
# Get general information on the dataframe of interest
df_target.info()
```

```python
# Drop rows with NA
df_target.dropna(inplace=True)
df_target.info()
```

```python
# Drop potential duplicated rows
df_target.drop_duplicates()
df_target.info()
```

```python
# Group mosted cited article (mca) by year to see the "popularity" of the journal
df_group_mca = df_target.set_index('Year_of_most_cited_article').groupby('Year_of_most_cited_article')
df_group_mca.head()
```

```python
# See count of the year of most cited article
df_mca_count = df_group_mca.count()
df_mca_count = df_mca_count.sort_values(by=['journal'],ascending=False)
Count_info = df_mca_count.journal
print(Count_info.head(15))
```

```python
# See mean of the most cited article citation 
df_mca_mean = df_group_mca.mean()
df_mca_mean = df_mca_mean.sort_values(by=['Citation_most_cited_article'],ascending=False)
Mean_info = df_mca_mean.Citation_most_cited_article
print(Mean_info.head(15))
```

```python
# See mean of the impact factor
df_mca_mean_ipf = df_mca_mean.sort_values(by=['impact_factor'],ascending=False)
Mean_ipf_info = df_mca_mean_ipf.impact_factor
print(Mean_ipf_info.head(15))
```

```python
# See mean of the article number
df_mca_mean_atn = df_mca_mean.sort_values(by=['article_number'],ascending=False)
Mean_atn_info = df_mca_mean_atn.article_number
print(Mean_atn_info.head(15))
```
